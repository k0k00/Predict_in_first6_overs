# Load Model:

```python
#from torch.autograd import Variable
class linearRegression(torch.nn.Module):
    def __init__(self, inputSize, outputSize):
        super(linearRegression, self).__init__()
        self.linear = torch.nn.Linear(inputSize, outputSize)

    def forward(self, x):
        out = self.linear(x)
        return out

```

```python
loaded_model = linearRegression(20,1)
loaded_model.load_state_dict(torch.load('iplscorePred.pth'))
loaded_model.eval()
```

# Normalize the data

```python
with open(Path.joinpath(Path.cwd().parent,'dataset','min_max_norm_df.json')) as f:
    mms_data = json.load(f)

mms_data
```

```python
input_data_norm = {}
for col in input_data:
    try:
        input_data_norm[col]  = (input_data[col] - mms_data[col]['min'])/(mms_data[col]['max'] - mms_data[col]['min'])
    except Exception as e:
        input_data_norm[col] = 0.0

input_data_norm
```

```python
predicted_score = loaded_model(torch.Tensor(np.array(list(input_data_norm.values()),dtype=float)))
predicted_score
```
